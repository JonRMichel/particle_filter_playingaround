```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
time_trend = 1.0

sigsq  = 1
sigsq_u = 1
sample_size = 50

state = [np.random.normal(time_trend, sigsq)]
Y = [state[-1] + np.random.normal(0.0, sigsq_u) ]

for time in range(1, sample_size):
    state.append( state[-1] + np.random.normal(time_trend, sigsq) )
    Y.append(state[-1] + np.random.normal(0.0, sigsq_u))
```

```python
fig, ax = plt.subplots(figsize = (18,6))
pd.DataFrame({'Observations':Y, 'state':state}).plot(ax = ax)
```

```python
pd.DataFrame({'Observations':Y, 'state':state}).diff(1).std()
```

```python
scipy.stats.norm(0, 1).pdf(0)
```

```python
particle_count = 1000
particle_df = pd.DataFrame(columns = range(0,particle_count))

particle_df.loc[0] = time_trend + np.random.uniform(1,2, particle_count)

time_trend_guess = 2.0
sigsq_ep_guess = 2.0
sigsq_u_guess = 2.0

for time in range(0, sample_size):
    weights = np.exp( (-1/(2*sigsq_u_guess))* np.square(Y[time] - particle_df.loc[time].values))
    new_particles =  random.choices(particle_df.loc[time].values , weights, k=particle_count) +  np.random.normal(0,sigsq_ep_guess, particle_count)
    particle_df.loc[time+1] = new_particles + time_trend_guess
```

```python
fig, ax = plt.subplots(figsize = (16,8))

pd.DataFrame({'Observations':Y, 'state':state}).plot(ax = ax, linewidth = 3)
particle_df.mean(axis = 1).plot(ax =ax , color = 'black')
particle_df.quantile(.9, axis =1).plot(color = 'gray',  linestyle = ':')
particle_df.quantile(.1, axis =1).plot(color = 'gray',  linestyle = ':')

ax.legend(['Observations','True State','Mean Particle State', '90th Percentile of Filter','10th Percentile of Filter'])
```

```python
print( particle_df.mean(axis = 1).diff(1).mean(), particle_df.mean(axis = 1).diff().std(), np.square(Y - particle_df.mean(axis = 1)[1:]).std())
```

```python
guess =[particle_df.mean(axis = 1).diff(1).mean(), particle_df.mean(axis = 1).diff().std(), np.square(Y - particle_df.mean(axis = 1)[1:]).std() ]
particle_count = 1000

for k in range(0,20):
    particle_df = pd.DataFrame(columns = range(0,particle_count))

    particle_df.loc[0] = time_trend + np.random.uniform(1,2, particle_count)

    time_trend_guess = guess[0]
    sigsq_ep_guess = guess[1]
    sigsq_u_guess = guess[2]

    for time in range(0, sample_size):
        weights = np.exp( (-1/(2*sigsq_u_guess))* np.square(Y[time] - particle_df.loc[time].values))
        new_particles =  random.choices(particle_df.loc[time].values , weights, k=particle_count) +  np.random.normal(0,sigsq_ep_guess, particle_count)
        particle_df.loc[time+1] = new_particles + time_trend_guess


    guess =  [particle_df.mean(axis = 1).diff(1).mean(), particle_df.mean(axis = 1).diff().std(), np.square(Y - particle_df.mean(axis = 1)[1:]).std() ]
    print(guess)
```

```python
particle_count = 100
min_obj = 1000
best_guess = [1,1,1]

for time_trend_guess in np.linspace(0,2,15):
    for sigsq_ep_guess in np.linspace(1,2,10):
        for sigsq_u_guess in np.linspace(0,1,10):
            
            guess = [time_trend_guess, sigsq_ep_guess, sigsq_u_guess]
            particle_df = pd.DataFrame(columns = range(0,particle_count))

            particle_df.loc[0] = time_trend + np.random.uniform(1,2, particle_count)

            
            for time in range(0, sample_size):
                weights = np.exp( (-1/(2*sigsq_u_guess))* np.square(Y[time] - particle_df.loc[time].values))
                new_particles =  random.choices(particle_df.loc[time].values , weights, k=particle_count) +  np.random.normal(0,sigsq_ep_guess, particle_count)
                particle_df.loc[time+1] = new_particles + time_trend_guess
                
            new_obj = np.square(Y - particle_df.mean(axis = 1)[1:]).std() 
            
            if new_obj < min_obj:
                min_obj = new_obj
                best_guess = guess
            
            
best_guess
```
